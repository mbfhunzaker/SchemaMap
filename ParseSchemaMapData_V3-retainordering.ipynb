```python
response_list = []
my_keys = set()
my_keys.add("ID")
for participant in participants[1:]:
    assoc_dict = {}
    associations = participant["Associations"].split(";")
    counter = 1
    for association in associations[:-1]:
        association = association.split(",")
        assoc_tuple = tuple(sorted([convert_keys(int(association[0]), convert_key_dict), convert_keys(int(association[1]), convert_key_dict)]))
        assoc_dict[assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_time"]= association[2]
        my_keys.add(assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_time")
        assoc_dict[assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_value"] =association[3]
        my_keys.add(assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_value")
        assoc_dict[assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_di"] =counter
        my_keys.add(assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_di")
        assoc_dict[assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_f"] = convert_keys(int(association[0]), convert_key_dict)
        my_keys.add(assoc_tuple[0].replace(" ", "")+"_"+assoc_tuple[1].replace(" ", "")+"_f")
        counter += 1
    assoc_dict["ID"] = participant["ID"]
    response_list.append(assoc_dict)

```

```python
#create function for clean words
def word_clean(word):
    lower = word.lower()
    return lower.strip(punctuation)

#create function for clean sentences
def text_clean(text):
    #split sentence
    words = text.split()
    #create list
    clean_words = []
    for word in words:
        w = word_clean(word)
        clean_words.append(w)
    return clean_words
```
